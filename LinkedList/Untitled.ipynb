```python
### Remove duplicates from the linked list

class Node:
    def __init__(self,data=0):
        self.val=data
        self.next=None
        
def make_list(elements):
    head=Node(elements[0])
    for element in elements[1:]:
        ptr=head
        while ptr.next:
            ptr=ptr.next
        ptr.next=Node(element)
    return head

def print_list(head):
    ptr=head
    print('[',end='')
    while ptr:
        print(ptr.val,end=',')
        ptr=ptr.next
    print(']')
    
class Solution:
    def deleteDuplicates(self,head):
        if not head:
            return
        cur = head.next
        start = prev = head
        while cur:
            if cur.val == prev.val:
                prev.next = cur.next
            else:
                prev = cur            
            cur = cur.next
        return start
    
ob = Solution()
head = make_list([2, 4, 6, 1, 4, 6, 9])
print_list(ob.deleteDuplicates(head))
```

```python
## Delete duplicates from the linkedlist

def deleteDuplicates(self, head: ListNode) -> ListNode:
    
        if head is None:
            return

        if head.next is not None:
    
            if head.val == head.next.val:
        
                temp = head.next.next
                head.next = None
                head.next = temp
                self.deleteDuplicates(head)
            else:
        
                self.deleteDuplicates(head.next)
        
        return head
```

```python
## Linked List cycle (Amazon, Microsoft, Sansung)

class Solution:
    def hasCycle(self,head):
        slow,fast=head,head
        while fast and fast.next:
            slow=slow.next
            fast=fast.next
            
            if slow==fast:
                return True
        return False
```

```python
## Intersection of two Linked List

class Solution:
    def getIntersection(self,headA,headB):
        p1=headA
        p2=headB
        
        while p1 != p2:
            if p1:
                p1=p1.next
            else:
                p1=headB
            if p2:
                p2=p2.next
            else:
                p2=headA
        return p1
```

```python
## rotate list/ Rotate from the right (FaceBook) (Medium)
'''I/P:1,2,3,4, O/p:4,5,1,2,3, k=2'''

class Solution:
    def rotateRight(self,head):
        ##checking if the head is not empty
        if not head:
            return head
        
        ## getting the length and tail
        length,tail=1,head
        ## iterating and updating the tail.next
        while tail.next:
            tail=tail.next
            length+=1
            
        ## making sure k is less tha the length og the linked list
        k=k%length
        if k==0:
            return head
        
        ## going towards pivot and rotate the list
        curr=head
        
        for i in range(length-k-1):
            curr=curr.next
        newHead=curr.next
        curr.next=None
        tail.next=head
        return newHead
```

```python
## Swap Node in Linked List (Medium)

''' Two Pointer Method '''

class Solution:
    def swapNodes(self, head: ListNode, k: int) -> ListNode:
	
	    # Initial State
        slow, fast = head, head
		
		# Phase 1
        for _ in range(k - 1):
            fast = fast.next
        first = fast

        # Phase 2
        while fast.next:
            slow, fast = slow.next, fast.next
		
		# Last
        first.val, slow.val = slow.val, first.val

        return head
```
