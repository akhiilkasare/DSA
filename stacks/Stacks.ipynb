```python
## Implement stack using queue

'''From scratch'''

import collections

class Stack:
    def __init__(self):
        self._queue=collections.deque()

    def push(self,x):
        q=self._queue
        q.append(x)

        for _ in range(len(q)-1):
            q.append(q.popleft())

    def pop(self):
        return self._queue.pop()

    def top(self):
        return self._queue[0]

    def empty(self):
        return not len(self._queue)
```

```python
## Valid Parenthesis (Facebook)(Easy)

class Solution:
    def isValid(self,s):
        stack=[]
        closeToOpen={')' : '(', ']' : '[', '}' : '{'}

        for c in s:
            if c in closeToOpen:
                if stack and stack[-1] == closeToOpen[c]:
                    stack.pop()
                else:
                    return False
            else:
                stack.append(c)

        return True if not stack else False
            
```

# Design a Min Stack (Amazon) (Easy)

'''
1. create a solution class
2. create a construtor and create 2 list i.e self.stack, self.minstack
3. push : append val in self.stack 
4. create a val variable and calculate the minimum of val and top element of the self.minstack
    if self.minstack else val
5. pop : remove element from both stacks i.e self.stack.pop() and self.minstack.pop()
6. top : get the top element of stack
7. getMin : get top on self.minstack
'''

class Solution:
    def __init__(self):
        self.stack=[]
        self.minstack=[]
    
    def push(self,val):
        self.stack.append(val)
        val = min(val, self.minstack[-1] if self.minstack else val)
        self.minstack.append(val)

    def pop(self):
        self.stack.pop()
        self.minstack.pop()

    def top(self):
        return self.stack[-1]

    def getMin(self):
        return self.minstack[-1]

```python
## Generate Parenthesis

'''
1. create 2 empty list i.e stack=[] res=[]
2. create a backtrack function with openN & closen
3. check if openN==closeN==n then append empty string in res and do .join(stack)
4. check if openN < n -> inside stack append one opening bracket i.e "("
5. and then call the backtrack function and inside it increament openN
6. then do stack.pop()
7. similarly for case if closeN < openN
8. then outside the if loop outside function call backtrack(0,0) and return res
'''

class Solution:
    def generateparenthesis(self,n):
        stack=[]
        res=[]

        def backtrack(openN, closeN):
            if openN == closeN == n:
                res.append("".join(stack))
                return 
            
            if openN < n:
                stack.append("(")
                backtrack(openN+1,closeN)
                stack.pop()

            if closedN < openN:
                stack.append(")")
                backtrack(openN,closeN+1)
                stack.pop()
        backtrack(0,0)
        return res  
```

```python
## Largest Rectangle in Histogram (Hard) (Amazon)

class Solution:
    def largestRectangleArea(self,heights):
        maxArea=0
        stack=[]

        for i,h in enumerate(heights):
            start=i
            while stack and stack[-1][1] > h:
                index,height=stack.pop()
                maxArea = max(maxArea, height*(i-index))
                start=index
            stack.append((start,h))
        
        for index,height in enumerate(heights):
            maxArea = max(maxArea, h*(len(heights)-i))
        return maxArea 
```

```python
## Reverse a Stack using Recursion

class Stack:
    def __init__(self):
        self.items = []

    def push(self,data):
        self.items.append(data)

    def pop(self):
        return self.items.pop()

    def isempty(self):
        return self.items == []

    def display(self):
        for data in reversed(self.items):
            print(data)

def insert_at_bottom(s,data):
    if s.isempty():
        s.push(data)
    else:
        popped = s.pop()
        insert_at_bottom(s,data)
        s.push(popped)

def reverse_stack(s):
    if not s.isempty():
        popped = s.pop()
        reverse_stack(s)
        insert_at_bottom(s,popped)

s = Stack()
data_list = input("Enter the values : ").split()
for data in data_list:
    s.push(int(data))

print("The original stack")
s.display()
reverse_stack(s)
print("After reversing : ")
s.display()


```

```python
## Sort a Stack usng Recursion

def sortInserted(s,data):
    if len(s) == 0 or data > s[-1]:
        s.append(data)
        return
    else:
        popped = s.pop()
        sortInserted(s,data)
        s.append(popped)
        
def sortStack(s):
    if len(s) != 0:
        popped = s.pop()
        sortStack(s)
        sortInserted(s,popped)

def printStack(s):
    for i in s[::-1]:
        print(i,end='')
    print()
    
if __name__ == "__main__":
    s = []
    s.append(30)
    s.append(-5)
    s.append(18)
    s.append(14)
    s.append(-3)
 
    print("Stack elements before sorting: ")
    printStack(s)
 
    sortStack(s)
 
    print("\nStack elements after sorting: ")
    printStack(s)
```

```python
# Return the largest rectangle area

class Solution:
    def largestArea(self, heights):
        maxArea=0
        stack=[]
        for i,h in enumerate(heights):
            start=i
            while stack and stack[-1][1] > h:
                index,height=stack.pop()
                maxArea=max(maxArea, height * (i-index))
                start = index
            stack.append((start,h))
        
        for i,h in stack:
            maxArea = max(maxArea, h * (len(heights) - i))
        return maxArea
```
